Importing the  Dependencies

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

Data Collection and preprocessing

```python
# loading the csv data to the pandas dataFrame
heart_data = pd.read_csv('/content/drive/MyDrive/Dataset/heart_disease_data.csv')
```

```python
# checking first 5 rows of the dataset
heart_data.head()
```

```python
# checking the no. of rows and columns of the dataset
heart_data.shape
```

```python
# getting some statical measure of the dataset
heart_data.describe()
```

```python
# getting some information about dataset
heart_data.info()
```

```python
# checking for missing values
heart_data.isnull().sum()
```

```python
# checking correlaion
correlation = heart_data.corr()
correlation
```

```python
# checking the distribution of the target variable
heart_data['target'].value_counts()
```

1 --> Defective Heart

2 --> Healthy Heart

Splitting the Feature and the Target

```python
X = heart_data.drop(columns ='target', axis=1)
Y = heart_data['target']
```

```python
print(X)
print("-----------------------------")
print(Y)
```

Splitting the data into training and the test data

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, train_size = 0.2, random_state =2)
```

```python
# printing the no. of train and test data
print(X.shape, X_train.shape, X_test.shape)
```

Model Training

Logistic Regression

```python
model = LogisticRegression()
```

```python
# trtaining the LogisticRegression with training data
model.fit(X_train, Y_train)
```

Model Evolution

Accuracy Score


```python
# accuracy on traininhg Data
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
```

```python
print("Accuracy Scorre : ", training_data_accuracy )
```

```python
# accuracy on test data
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
```

```python
print("Accuray on Test Data : ", test_data_accuracy)
```

Building a Predictive System

```python
input_data = (62,0,0,140,268,0,0,160,0,3.6,0,2,2)

# changing input into numpy array
input_data_as_numpy_array = np.asarray(input_data)

#reshape the numpy array as we are predicting for only on instance
input_data_reshaped = input_data_as_numpy_array.reshape(1, -1)

prediction = model.predict(input_data_reshaped)
print(prediction)

if (prediction[0]==0):
  print("The Person does not have a heart disease")

else:
  print("The Person has Heart Disease")

```
