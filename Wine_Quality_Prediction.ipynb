```python
from os import setns
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
```

importing dataset

```python
# loading the dataset to a pandas dataframe
wine_dataset =pd.read_csv('/content/drive/MyDrive/Dataset/winequality-red.csv')
wine_dataset.head()
```

Data Preprocessing

```python
# seperate the data and the Label
X = wine_dataset.drop('quality', axis = 1)
```

Label Binarization

```python
Y = wine_dataset['quality'].apply(lambda y_value:1 if y_value>=7 else 0)
```

```python
Y.unique()
```

Train & Test split

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size= 0.2, random_state =2 )
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

Model Training

Random Forest Classifier

```python
model = RandomForestClassifier()
```

```python
model.fit(X_train, Y_train)
```

Model Evaluation

Accuracy Score

```python
# accuracy on training data
X_train_prediction = model.predict(X_train)
train_data_accuracy = accuracy_score(X_train_prediction, Y_train)
print(train_data_accuracy)
```

```python
# accuracy on test data
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
print(test_data_accuracy)
```

Building a predictive system

```python
input_data = (7.5,0.5,0.36,6.1,0.071,17.0,102.0,0.9978,3.35,0.8,10.5)

# changing the inpurt data into a numpy array
input_data_as_numpy_array = np.asarray(input_data)

# reshape the data as we are predicting the label for only onr instance
input_data_reshaped = input_data_as_numpy_array.reshape(1, -1)

prediction = model.predict(input_data_reshaped)
print(prediction)

if (prediction[0]==1):
  print('Good Quality Wine')

else:print('Bad Quality Wine')
```
